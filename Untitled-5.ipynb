```python
# =============================================================
# ITERATION 1 вҖ” FINAL WORKING ENHANCED CNN
# =============================================================

import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv1D, MaxPooling1D, Dense, Dropout,
    BatchNormalization, GlobalAveragePooling1D
)
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping


# =============================================================
# PATHS
# =============================================================
TRAIN_PATH = '/Users/gracytanna/Downloads/processed_train_vowel_and_cough.csv'
VAL_PATH   = '/Users/gracytanna/Downloads/processed_val_vowel_and_cough.csv'
TEST_PATH  = '/Users/gracytanna/Downloads/processed_test_vowel_and_cough.csv'

LABEL_COL = 'disease'
CLASS_NAMES = ['Healthy', 'Asthma', 'COPD', 'COVID']

EPOCHS = 100
BATCH_SIZE = 32
PATIENCE = 10
RANDOM_SEED = 42

np.random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)


# =============================================================
# COLUMN NORMALIZATION
# =============================================================
def normalize_col(c):
    c = re.sub(r'-cough$', '', c)
    c = re.sub(r'-vowel$', '_vowel', c)
    return c


# =============================================================
# LOAD DATA
# =============================================================
print("рҹ“Ӯ Loading data...")

train_df = pd.read_csv(TRAIN_PATH).rename(columns=normalize_col)
val_df   = pd.read_csv(VAL_PATH).rename(columns=normalize_col)
test_df  = pd.read_csv(TEST_PATH).rename(columns=normalize_col)

print(f"Train: {train_df.shape} | Val: {val_df.shape} | Test: {test_df.shape}")


# =============================================================
# FEATURE SELECTION
# =============================================================
FEATURE_COLS = train_df.select_dtypes(include=[np.number]).columns.tolist()

if LABEL_COL in FEATURE_COLS:
    FEATURE_COLS.remove(LABEL_COL)

print(f"вң… Features: {len(FEATURE_COLS)}")


# =============================================================
# SPLIT + ALIGN
# =============================================================
X_train_raw = train_df[FEATURE_COLS]
y_train_raw = train_df[LABEL_COL]

X_val_raw   = val_df[FEATURE_COLS]
y_val_raw   = val_df[LABEL_COL]

X_test_raw  = test_df[FEATURE_COLS]
y_test_raw  = test_df[LABEL_COL]


# =============================================================
# HANDLE NON-NUMERIC + NaNs
# =============================================================
X_train_raw = X_train_raw.apply(pd.to_numeric, errors='coerce').fillna(0)
X_val_raw   = X_val_raw.apply(pd.to_numeric, errors='coerce').fillna(0)
X_test_raw  = X_test_raw.apply(pd.to_numeric, errors='coerce').fillna(0)


# =============================================================
# SCALING
# =============================================================
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train_raw)
X_val   = scaler.transform(X_val_raw)
X_test  = scaler.transform(X_test_raw)


# =============================================================
# LABEL ENCODING
# =============================================================
le = LabelEncoder()
le.fit(pd.concat([y_train_raw, y_val_raw, y_test_raw]))

y_train = le.transform(y_train_raw)
y_val   = le.transform(y_val_raw)
y_test  = le.transform(y_test_raw)

num_classes = len(le.classes_)

y_train_oh = to_categorical(y_train, num_classes)
y_val_oh   = to_categorical(y_val, num_classes)


# =============================================================
# CLASS WEIGHTS
# =============================================================
cw = compute_class_weight('balanced',
                          classes=np.unique(y_train),
                          y=y_train)

class_weight_dict = dict(enumerate(cw))
print(f"вң… Class weights: {class_weight_dict}")


# =============================================================
# RESHAPE FOR CNN
# =============================================================
n_features = X_train.shape[1]

X_train_cnn = X_train.reshape(-1, n_features, 1)
X_val_cnn   = X_val.reshape(-1, n_features, 1)
X_test_cnn  = X_test.reshape(-1, n_features, 1)


# =============================================================
# MODEL
# =============================================================
model = Sequential([
    Conv1D(64, 3, activation='relu', padding='same',
           input_shape=(n_features, 1)),
    BatchNormalization(), MaxPooling1D(2), Dropout(0.3),

    Conv1D(128, 3, activation='relu', padding='same'),
    BatchNormalization(), MaxPooling1D(2), Dropout(0.3),

    Conv1D(256, 3, activation='relu', padding='same'),
    BatchNormalization(),
    GlobalAveragePooling1D(),

    Dense(128, activation='relu'), Dropout(0.4),
    Dense(64, activation='relu'), Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()


# =============================================================
# TRAIN
# =============================================================
history = model.fit(
    X_train_cnn, y_train_oh,
    validation_data=(X_val_cnn, y_val_oh),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    class_weight=class_weight_dict,
    callbacks=[EarlyStopping(patience=PATIENCE,
                             restore_best_weights=True)]
)


# =============================================================
# EVALUATE
# =============================================================
y_pred = np.argmax(model.predict(X_test_cnn), axis=1)

print("\nрҹ“Ҡ Classification Report:\n")
print(classification_report(
    y_test,
    y_pred,
    labels=[0,1,2,3],
    target_names=CLASS_NAMES
))


# =============================================================
# CONFUSION MATRIX
# =============================================================
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=CLASS_NAMES,
            yticklabels=CLASS_NAMES,
            cmap='Blues')

plt.title("Enhanced CNN Confusion Matrix")
plt.tight_layout()
plt.show()


# =============================================================
# TRAINING CURVES
# =============================================================
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.legend(); plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.legend(); plt.title('Loss')

plt.tight_layout()
plt.show()


print("\nвң… DONE вҖ” ITERATION 1 WORKS PERFECTLY")
```
